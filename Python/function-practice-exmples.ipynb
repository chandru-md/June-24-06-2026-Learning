```python
def temperature_conversion(temp,unit):
    ## This functions converts celcius to fahrenheit and vice-versa
    if unit == "C":
        return temp * 9/5 + 32
    elif unit == "F":
        return temp - 32 * 5/9
    else:
        return None
```

```python
temperature_conversion(44,"C")
```

```python
temperature_conversion(32,"F")
```
