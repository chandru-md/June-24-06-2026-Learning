```python
## Temperature conversion
```

```python
def temperature_conversion(temp,unit):
    ## This functions converts celcius to fahrenheit and vice-versa
    if unit == "C":
        return temp * 9/5 + 32
    elif unit == "F":
        return temp - 32 * 5/9
    else:
        return None
```

```python
temperature_conversion(44,"C")
```

```python
temperature_conversion(32,"F")
```

```python
## Password Strength checker
```

```python
def is_strong_password(password):
    ## This function checks the password is strong or not

    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True
```

```python
is_strong_password("Kathiravan@#$1234")
```

```python
is_strong_password("Girl")
```

```python
## Total cost calculator
```

```python
def total_cost_calculator(cart):
    total_cost = 0
    for item in cart:
        total_cost = item["price"] + item["quantity"]
    return total_cost

cart = [
    {"name":"Apple","price":4.5,"quantity":6},
    {"name":"Beetroot",'price':3.2,'quantity':8},
    {'name':"Carror",'price':2.8,"quantity":5}
]
```

```python
total_cost =total_cost_calculator(cart)

print(total_cost)
```

```python
## Check if a string is a palindrome
```

```python
 # Palindrome
```

```python
## aba  === aba
```

```python
def is_palindrome(s):
    s = s.lower().replace(" ","")
    return s==s[::-1]
```

```python
is_palindrome("A man a plan a canal Panama")
```

```python
is_palindrome("A man a plan string canal ")
```

```python
## Calculate the factorials of a number using recursion
```

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(5)
```

```python
## Sample.txt

## A function to read a file and count the frequency of each word
```

```python
def count_word_frequency(file_path):
    word_count = {}
    with open(file_path,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('!@#$%^&*()')
                word_count[word] = word_count.get(word,0)+1
    return word_count

filepath = "sample.txt"
word_frequency = count_word_frequency(filepath)
print(word_frequency)
```
