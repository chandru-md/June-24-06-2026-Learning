```python
## Temperature conversion
```

```python
def temperature_conversion(temp,unit):
    ## This functions converts celcius to fahrenheit and vice-versa
    if unit == "C":
        return temp * 9/5 + 32
    elif unit == "F":
        return temp - 32 * 5/9
    else:
        return None
```

```python
temperature_conversion(44,"C")
```

```python
temperature_conversion(32,"F")
```

```python
## Password Strength checker
```

```python
def is_strong_password(password):
    ## This function checks the password is strong or not

    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True
```

```python
is_strong_password("Kathiravan@#$1234")
```

```python
is_strong_password("Girl")
```

```python
## Total cost calculator
```

```python
def total_cost_calculator(cart):
    total_cost = 0
    for item in cart:
        total_cost = item["price"] + item["quantity"]
    return total_cost

cart = [
    {"name":"Apple","price":4.5,"quantity":6},
    {"name":"Beetroot",'price':3.2,'quantity':8},
    {'name':"Carror",'price':2.8,"quantity":5}
]
```

```python
total_cost =total_cost_calculator(cart)

print(total_cost)
```
