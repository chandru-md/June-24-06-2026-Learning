```python
## MAp Functions
```

```python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x**2, numbers))

print(result)
```

```python
nums = ["10", "20", "30", "40"]

result = list(map(int, nums))

print(result)
```

```python
numbers = [1, 2, 3, 4]

result = list(map(str, numbers))

print(result)
```

```python
names = ["alice", "bob", "charlie"]

result = list(map(str.upper, names))

print(result)
```

```python
words = ["Python", "Java", "AI"]

result = list(map(len, words))

print(result)
```

```python
a = [1, 2, 3]
b = [10, 20, 30]

result = list(map(lambda x, y: x + y, a, b))

print(result)
```

```python
a = [2, 3, 4]
b = [5, 6, 7]

result = list(map(lambda x, y: x * y, a, b))

print(result)
```

```python
celsius = [0, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
```

```python
names = [" Alice ", " Bob ", " Charlie "]

result = list(map(str.strip, names))

print(result)
```

```python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x % 2 == 0, numbers))

print(result)
```

```python
def greet(name):
    return f"Hello, {name}"

names = ["Alice", "Bob", "Charlie"]

result = list(map(greet, names))

print(result)
```

```python
marks = [75, 80, 65, 90]

updated = list(map(lambda x: x + 5, marks))

print(updated)
```

```python
cities = ["chennai", "delhi", "mumbai"]

result = list(map(str.capitalize, cities))

print(result)
```

```python
nums = [3.14159, 2.71828, 1.23456]

result = list(map(lambda x: round(x, 2), nums))

print(result)
```

```python
prices = [100, 250, 500]

final_prices = list(map(lambda p: p * 1.18, prices))

print(final_prices)
```
