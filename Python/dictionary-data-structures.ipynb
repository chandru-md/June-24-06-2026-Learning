```python
### Nested dictionaries
```

```python
students = {
    "student_1":{"name":"Devi Bala","age":25},
    "student_2":{"name":"Sweety Omana","age":25}
}
```

```python
students
```

```python
print(students)
```

```python
print(students["student_1"]["name"])
```

```python
print(students["student_2"]["name"])
```

```python
for id,info in students.items():
    print(f"{id}:{info}")
```

```python
for id,info in students.items():
    # print(f"{id}:{info}")
    for key,value in info.items():
        print(f"{key}:{value}")
```

```python
# Dictionary comprehension
```

```python
squares = {x:x**2 for x in range(5)}

print(squares)
```

```python
## Conditional dictionary comprehension
```

```python
squares = {x:x**2 for x in range(10) if x%2==0}

print(squares)
```

```python
## Frequency count
```

```python
numbers = {1,2,2,3,3,3,4,4,4,4}

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] = frequency[number] + 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
## merge dictionaries
```

```python
d1 = {'a':1,'b':2}

d2 = {'b':3,"c":4}

merge_dict = {**d1,**d2}

print(merge_dict)
```
