```python
# Data-types - in python
```

```python
# Integer Example

age = 25
print(age)

print(type(age))
```

```python
# Float Example

distance = 31.44
print(distance)

print(type(distance))
```

```python
# Boolean Example

is_employee = True
print(is_employee)

print(type(is_employee))
```

```python
# String Example

name = "Sweety Omana"

print(name)

print(type(name))
```

```python
# Common Errors

name = "Chandru" + 85

print(name)
```

```python

name = "Chandru" + str(85)

print(name)
```
