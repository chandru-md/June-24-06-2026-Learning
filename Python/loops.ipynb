```python
## While loop
```

```python
## the while loops continues to execute as long as the condition is true.
```

```python
count = 0

while count < 11:
    print(count)
    count += 1
```

```python
count = 0

while count %2 ==0:
    print(count)
    count += 1
```

```python
## loop control statements
```

```python
## break

## continue

## pass

## None
```

```python
for i in range(0,11):
    if i == 5:
        break
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(5):
    pass
```

```python
## Nested loops
```

```python
# A loop inside a loop
```

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## eg. cal the sum of first n natural numbers using while and for loop.
```

```python
n = 10

sum = 0

count = 1

while count <=n:
    sum = sum + count
    count +=1


print("The first 10 Natural Numbers are ",sum)
```

```python
# eg.. prime numbers b/w 1 and 100
```

```python
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
