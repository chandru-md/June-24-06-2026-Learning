```python
## While loop
```

```python
## the while loops continues to execute as long as the condition is true.
```

```python
count = 0

while count < 11:
    print(count)
    count += 1
```

```python
count = 0

while count %2 ==0:
    print(count)
    count += 1
```

```python
## loop control statements
```

```python
## break

## continue

## pass

## None
```

```python
for i in range(0,11):
    if i == 5:
        break
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(5):
    pass
```
