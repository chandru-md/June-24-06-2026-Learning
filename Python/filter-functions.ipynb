```python
## Filter Functions
```

```python
numbers = [1, 2, 3, 4, 5, 6]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```

```python
numbers = [1, 2, 3, 4, 5, 6]

result = list(filter(lambda x: x % 2 != 0, numbers))

print(result)
```

```python
marks = [35, 60, 45, 80, 90, 25]

result = list(filter(lambda x: x > 50, marks))

print(result)
```

```python
numbers = [-5, 3, -1, 8, 0, 10]

result = list(filter(lambda x: x > 0, numbers))

print(result)
```

```python
names = ["Alice", "Bob", "Andrew", "Charlie", "Anita"]

result = list(filter(lambda name: name.startswith("A"), names))

print(result)
```

```python
words = ["AI", "Python", "Java", "Machine", "SQL"]

result = list(filter(lambda word: len(word) > 5, words))

print(result)
```

```python
items = ["Python", "", "AI", "", "Data Science"]

result = list(filter(None, items))

print(result)
```

```python
letters = ['a', 'b', 'c', 'e', 'i', 'o', 'u']

vowels = list(filter(lambda ch: ch in "aeiou", letters))

print(vowels)
```

```python
numbers = [3, 5, 6, 9, 10, 12, 14]

result = list(filter(lambda x: x % 3 == 0, numbers))

print(result)
```

```python
def is_adult(age):
    return age >= 18

ages = [12, 17, 18, 21, 15, 30]

adults = list(filter(is_adult, ages))

print(adults)
```

```python
marks = [35, 48, 67, 80, 29, 55]

passed = list(filter(lambda x: x >= 50, marks))

print(passed)
```

```python
files = ["main.py", "notes.txt", "app.py", "image.png"]

python_files = list(filter(lambda f: f.endswith(".py"), files))

print(python_files)
```

```python
products = {
    "Laptop": 75000,
    "Mouse": 800,
    "Keyboard": 2500,
    "Pen": 20
}

expensive = list(filter(lambda key: products[key] > 1000, products))

print(expensive)
```

```python
words = ["madam", "python", "level", "radar", "hello"]

palindromes = list(filter(lambda w: w == w[::-1], words))

print(palindromes)
```

```python
prices = [450, 1200, 799, 2500, 999]

affordable = list(filter(lambda price: price < 1000, prices))

print(affordable)
```
