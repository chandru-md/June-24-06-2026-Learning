```python
text = input("Enter a string")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
def is_palindrome(text):
    text = text.lower()
    return text == text[::-1]

word = input("Enter the word: ")

if is_palindrome(word):
    print(f"{word} is a palindrome")
else:
    print(f"{word} is not a palindrome")

```

```python
text = input("enter a String: ")

reverse = ""

for char in text:
    reverse = char + reverse

if text == reverse:
    print("Palindrome")
else:
    print("Not a palindrome")
```
