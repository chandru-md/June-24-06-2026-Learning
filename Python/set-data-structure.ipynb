```python
# mathematical Operations
```

```python
set1 = {1,2,3,4,5,6}

set2 = {4,5,6,7,8,9}
```

```python
# Union

union_set = set1.union(set2)

print(union_set)
```

```python
# union = combining everything
```

```python
## intersection
```

```python
intersection_set = set1.intersection(set2)

print(intersection_set)
```

```python
# intersection gives common elements
```

```python
set1.intersection_update(set2)

print(set1)
```

```python
set1 = {1,2,3,4,5,6}
```

```python
set2 ={4,5,6,7,8,9}
```

```python
# Difference


print(set1.difference(set2))
```

```python
# difference = set1.difference(set2) - common elements removed from the set1
```

```python
print(set1.difference_update(set2))

print(set1)
```

```python
# Symmetric DIfference
```

```python
print(set1.symmetric_difference(set2))
```

```python
## symmetric difference = is common elements will be removed and  unique elements are combined.
```
